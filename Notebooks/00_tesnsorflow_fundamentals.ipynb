What we have created so far

*Scalar: a single number

*Vector: a number with direction (e.g. wind speed and direction)

*Matrix: a 2-dimensional array of numbers

*Tensor: a n-dimensional array of numbers where n can be any number, a 0-dimensional tensor is a scalar, 1-dimensional is a vector.

##Creating random tensors

Random tensors are tensor of some arbitary size which contain some random numbers



```python
# Creating two random (but the same) tensors

random_1 = tf.random.Generator.from_seed(42)# set seed for reproducibility
random_1 = random_1.normal(shape=(3,2))
random_2 =tf.random.Generator.from_seed(42)
random_2 = random_2.normal(shape=(3,2))

random_1,random_2,random_1==random_2
```

### Shuffle the order of the elements in a tensor

```python
# shuffle a tensor (valuable for shuffling one's data)

not_shuffled = tf.constant([[10,7],[3,4],[2,5]])

# Shuffled our non-shuffled tensor
tf.random.set_seed(42)
tf.random.shuffle(not_shuffled,42)
```

```python
unshuffled_1 = tf.constant([[1,2],[3,4],[5,6]])
unshuffled_2 = tf.constant([[17,32],[39,47],[57,16]])
unshuffled_3 = tf.constant([[12,21],[33,45],[5,16]])
unshuffled_4 = tf.constant([[12,1],[33,5],[5,116]])

tf.random.set_seed(2)
tf.random.shuffle(unshuffled_4,seed=2)
```

### Getting info from tensors

Tensor Attributes should be familiared with:

* Shape
* Rank
* Axis or dimension
* Size

```python
# Create a rank 4 tensor

rank_4_tensor = tf.zeros(shape=(2,3,4,5))
rank_4_tensor
```

```python
rank_4_tensor.shape,rank_4_tensor.ndim,tf.size(rank_4_tensor)
```

```python
# Get various attributes of tensors

print("Datatype of every element:", rank_4_tensor.dtype)
print("Number of dimensions:", rank_4_tensor.ndim)
print("Shape of tensor:", rank_4_tensor.shape)
print("Elements along the 0 Axis:", rank_4_tensor.shape[0])
print("Elements along the last Axis:", rank_4_tensor.shape[-1])
print("Total number of elements:", tf.size(rank_4_tensor).numpy())
```

### Indexing tensors

Tensors can be indexed like python lists

```python
# Get the first two elements of each dimension

rank_4_tensor[:2,:2,:2,:2]
```

```python
# Get the first element from each dimension from each index except for the final one
rank_4_tensor[:1,:1,:1]
```

```python
# Create a rank 2 tensor (2 dimension)
rank_2_tensor = tf.constant([[10,7],[3,4]])
rank_2_tensor.ndim, rank_2_tensor.shape
```

```python
rank_2_tensor[:,-1]
```

```python
# Add in extra dimension to our rank 2 tensor

rank_3_tensor = rank_2_tensor[...,tf.newaxis]
rank_3_tensor
```

```python
# Alternative to tf.newaxis
tf.expand_dims(rank_2_tensor,axis=-1) 

# axis=-1 means expand the final axis
# axis=0 means expand the 0th axis
```

### Manipulating tensors (tensor operations)

**Basic operation**

`+`,  `-`,  `*`,  `/`

```python
# You can add values to a tensor using addition operator

tensor = tf.constant([[10,7],[3,4]])
tensor + 10
```

```python
# Multiplication of tensors
tensor*10
```

```python
# Suntraction of tensors
tensor-10
```

```python
# We can use the tensorflow built-in functions
tf.multiply(tensor,10)
```

**Matrix Multiplication**

In machine learning matrix multiplication is one of the most common operations

```python
# Matrix multiplication in tensorflow

tf.matmul(tensor,tensor)
```

```python
# Matrix multiplication with Python operator "@"

tensor @ tensor
```

```python
# Create a tensor of (3,2)
X = tf.constant([[1,2],[3,4],[5,6]])
# Create a tensor (2,3)
Y = tf.constant([[7,8,1],[9,4,6]])
X @ Y
```

```python
tf.transpose(X)
```

```python
# Perform the dot product on X and Y (requires X or Y to be transposed)
tf.tensordot(X,Y,axes=1)
```

### Changing the datatype of a tensor

```python
# Create a new tensor with default datatype (float32)
B = tf.constant([1.7,7.4])
B.dtype
```

```python
C = tf.constant([7,10])
C.dtype
```

```python
# Changing tensor from float32 to float16 (reduced precision)
D=tf.cast(B,dtype=tf.float16)
D, D.dtype
```

```python
# Change from int32 to float32
E = tf.cast(C, dtype=tf.float32)
E, E.dtype
```

### Aggregating Tensors

Aggregating tensors = condensing them from multiple values down to a smaller amount of values

```python
# Get the absolute values

D = tf.constant([-7,-10])
D
```

```python
tf.abs(D)
```

Let's go through the following form of aggregation:

* Get the minimum
* Get the maximum
* Get the mean of a tensor
* Get the sum of a tensor

```python
# Create a random tensor with values between 0 and 100 of size 50

E = tf.constant(np.random.randint(0,100,size=50))
E
```

```python
# Find the minimum
tf.reduce_min(E)
```

```python
# Find the maximum
tf.reduce_max(E)
```

```python
# Find the mean
tf.reduce_mean(E)
```

```python
# Find the sum
tf.reduce_sum(E)
```

```python
# Find the standard variance of a tensor
tf.math.reduce_std(tf.cast(E,dtype=tf.float32))
```

```python
# Find the variance of tensor
tf.math.reduce_variance(tf.cast(E,dtype=tf.float32))
```

### Find the positional maximum and minimum of a tensor

```python
tf.math.argmax(E)
```

```python
tf.math.argmin(E)
```

### One-hot encoding in tensor
